```python
import pandas as pd
import numpy as np
data = pd.read_csv('../Dataset/loan_prediction.csv')
data.head()
```

```python
data['Gender']=data['Gender'].map({'Female':1, 'Male':0}).fillna(data['Gender'])
data['Property_Area']=data['Property_Area'].map({'Urban':2,'Semiurban': 1,'Rural':0}).fillna(data['Property_Area'])
data['Married']=data['Married'].map({'Yes':1,'No':0}).fillna(data['Married'])
data['Education']=data['Education'].map({'Graduate':1,'Not Graduate':0}).fillna(data['Education'])
data['Self_Employed']=data['Self_Employed'].map({'Yes':1,'No':0}).fillna(data['Self_Employed'])
data['Loan_Status']=data['Loan_Status'].map({'Y':1,'N':0}).fillna(data['Loan_Status'])
```

```python
#Handling categorical feature Gender
data['Gender']=data['Gender'].map({'Female':1, 'Male':0}).fillna(data['Gender'])
data.head()
```

```python
data['Property_Area']=data['Property_Area'].map({'Urban':2, 'Semiurban': 1, 'Rural':0}).fillna(data['Property_Area'])
```

## Handling Missing values

```python
#finding the sum of null values in each column
data.isnull().sum()
```

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
```

```python
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
```

```python
#replacing + with space for filling the nan values
data['Dependents']=data['Dependents'].str.replace('+','')
```

```python
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
```

```python
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mode()[0])
```

```python
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0])
```

```python
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode()[0])
```

```python
data.isnull().sum()
```

```python
#getting bthe total info of the data after perfroming categorical to numericsal and rplacing misssing values
data.info()
```

```python
#changing the datatype of each float column to int
data['Gender']=data['Gender'].astype('int64')
data['Married']=data['Married'].astype('int64')
data['Dependents']=data['Dependents'].astype('int64')
data['Self_Employed']=data['Self_Employed'].astype('int64')
data['CoapplicantIncome']=data['CoapplicantIncome'].astype('int64')
data['LoanAmount']=data['LoanAmount'].astype('int64')
data['Loan_Amount_Term']=data['Loan_Amount_Term'].astype('int64')
data['Credit_History']=data['Credit_History'].astype('int64')
```

```python
data.info()
```

```python
#split data into features and target
x = data.drop(columns=['Loan_ID', 'Loan_Status'])
y = data['Loan_Status']
```

```python
#import and instantiate SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
#creating a new x and y varialbles for the balnced set
x_bal,y_bal = smote.fit_resample(x,y)
```

```python
#printing the values of y before balancing the data and after
print(y.value_counts())
print(y_bal.value_counts())
```

```python
names = x_bal.columns
```

## Scaling the dataset

```python
# performing feature Scaling operation using standard scaler on X part of the dataset because
# there different type of values in the columns
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_bal = sc.fit_transform(x_bal)
```

```python
x_bal = pd.DataFrame(x_bal, columns=names)
```

## Train Test Split

```python
# splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_bal, y_bal, test_size=0.33, random_state=42)
```

## Model Training and Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Instantiate models
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
gb = GradientBoostingClassifier(random_state=42)

models = {
    "Decision Tree": dt,
    "Random Forest": rf,
    "KNN": knn,
    "Gradient Boosting": gb
}

# Train and evaluate models
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"=== {name} ===")
    print(f"Accuracy: {acc:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("-" * 30)
```
